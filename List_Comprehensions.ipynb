```python
#to check whether 6 is present or not
n=int(input("Enter a number "))
s=str(n)
if '6' in s:
  print('6 is present')
else:
  print('6 is not present')
```

```python
#2 count number of 6 between 1-1000
n=int(input("Enter the range"))
L=[i for i in range(1,n+1) if "6" in str(i)]
#print(len(L))
res=0
for x in L:
  res=str(x).count("6")+res
print("The number of 6 present :",res)
```
