Lists need not be homogeneous always which makes it a most powerful tool in Python A single list may contain DataTypes like Integers, Strings, as well as Objects Lists are mutable, and hence, they can be altered even after their creation. Lists in Python can be created by just placing the sequence inside the square brackets[].
